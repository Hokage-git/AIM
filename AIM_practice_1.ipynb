```python
import pandas as pd

df = pd.read_csv("titanic.csv")
df
```

```python
df_female = df[df['Sex'] == 'female']
df_male = df[df['Sex'] == 'male']
df_male,df_female
```

```python
df_cleaned = df.dropna()
df_male = df_male.dropna()
df_female = df_female.dropna()
df_cleaned
```

```python
import matplotlib.pyplot as plt

df['Age'].plot(kind='hist', bins=100, alpha=0.7)
plt.xlabel('Возраст')
plt.ylabel('Количество пассажиров')
plt.grid(axis='y')
```

```python
def compute_statistics(dataframe):
    return {
        'min': dataframe['Age'].min(),
        'max': dataframe['Age'].max(),
        'mean': dataframe['Age'].mean(),
        'mode': dataframe['Age'].mode()[0],
        'median': dataframe['Age'].median()
    }
male_stats = compute_statistics(df_male)
female_stats = compute_statistics(df_female)

stats_df = pd.DataFrame({
    'Male': male_stats,
    'Female': female_stats
})
```

```python
df_male['Age'].plot(kind='line', label='Мужчины', color='blue', linestyle='-')
df_female['Age'].plot(kind='line', label='Женщины', color='pink', linestyle='-')

plt.title('Линейный график возраста')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```

```python
df_male['Age'].plot(kind='hist', alpha=0.5, label='Мужчины', color='blue', bins=10)
df_female['Age'].plot(kind='hist', alpha=0.5, label='Женщины', color='pink', bins=10)

plt.title('Гистограмма по возрасту')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df_male['Age'].plot(kind='box', label='Мужчины', color='blue')
df_female['Age'].plot(kind='box', label='Женщины', color='pink')

plt.title('Коробчатая диаграмма возраста')
plt.ylabel('Возраст')
plt.show()
```

```python
df_male['Age'].plot(kind='kde', label='Мужчины', color='blue', linestyle='--')
df_female['Age'].plot(kind='kde', label='Женщины', color='pink', linestyle='-.')

plt.title('График плотности по возрасту')
plt.xlabel('Возраст')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

```python
plt.scatter(df_male.index, df_male['Age'], label='Мужчины', color='blue', alpha=0.6)
plt.scatter(df_female.index, df_female['Age'], label='Женщины', color='pink', alpha=0.6)

plt.title('Точечная диаграмма по возрасту')
plt.ylabel('Возраст')
plt.legend()
plt.show()
```
